```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

Answer To The Q. No. 01

```python
import numpy as np
import pandas as pd
df=pd.read_csv("employees.csv")
df
```

```python
age_arr = df['age'].values
salary_arr = df['salary'].values

print("Age Array:")
print(age_arr)
print("dtype:", age_arr.dtype)
print("ndim:", age_arr.ndim)
print("shape:", age_arr.shape)
print("size:", age_arr.size)
print("\nSalary Array:")


print(salary_arr)
print("dtype:", salary_arr.dtype)
print("ndim:", salary_arr.ndim)
print("shape:", salary_arr.shape)
print("size:", salary_arr.size)
```

Answer To The Q. No. 2

```python
highest_salary=np.max(salary_arr)
print("Highest Salary:",highest_salary)

lowest_salary=np.min(salary_arr)
print("Lowest Salary:",lowest_salary)

age_avg=np.mean(age_arr)
print("Average Age:",age_avg)

salary_avg=np.mean(salary_arr)
print("Average Salary:",salary_avg)
```

Answer To The Q. No. 3

```python
ages_grater_than_30 = age_arr[age_arr > 30]
print("Ages grater than 30:")
print(ages_grater_than_30)

number_age_grater_than_30 = len(ages_grater_than_30)
print("\nNumber of employees older than 30:", number_age_grater_than_30)
```
